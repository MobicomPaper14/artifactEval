```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the uploaded Excel file to check its contents
file_path = 'CAD_data.xlsx'

# Load the data to inspect the structure
data = pd.read_excel(file_path)

# Display the first few rows of the data to understand the structure
data.head()
# Extracting relevant data
time = data['Time (sec)']
cad_values = data['CAD']
packet_decoded = data['Packet decoded']

# Create the plot
plt.figure(figsize=(10, 6))

# Plot CAD values over time as dots
plt.scatter(time, cad_values, color='blue', label='CAD', s=10)

# Plot Packet decoded as 'X' only when true (i.e., not NaN or 1)
plt.scatter(time[packet_decoded == "Yes"], [1] * sum(packet_decoded == "Yes"), 
            color='red', label='Packet decoded', marker='x', s=300)

# Adding labels and title
plt.xlabel('Time (sec)')
plt.ylabel('CAD Value')
plt.title('CAD Values Over Time with Received Packets Marked')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
